```python
import random
```

```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
emoji_dict = {
    "positive": ["😊", "😄", "👍", "🎉", "😍", "😁", "😎", "🤩", "💖", "👏"],
    "negative": ["😞", "😠", "👎", "😢", "😭", "💔", "😡", "😔", "😤", "😣"],
    "neutral":  ["😐", "🤔", "😶", "😑", "🙄", "🧐", "😬", "😕", "😒", "😴"]
}

keyword_emoji = {
     "happy": ["😄", "😊", "🥰"],
      "love": ["❤", "😍", "😘"],
      "angry": ["😠", "😡", "🤬"],
     "sad": ["😢", "😭", "😞"],
     "fun": ["🤣", "😜", "🎉"],
     "excited": ["🤩","🥳","😃"]
     }
```

```python
def gen_emojis(text):

  # get sentiment prediction
  result = sentiment_pipeline(text)[0]
  label = result['label']

# get emozi from sentiment
  sentiment_emojis = random.sample(emoji_dict.get(label, emoji_dict["NEUTRAL"]), 3)

# get keyword based emoji
  keyword_emojis=[]
  for keyword in keyword_emoji:
    if keyword in text.lower():
      keyword_emojis.extend(keyword_emoji[keyword])

  all_emojis = list(dict.fromkeys(sentiment_emojis + keyword_emojis))
  return ''.join(all_emojis[:3])
```

```python
input_text= input("Enter your mood or sentence: ")
emoji= gen_emojis(input_text)
print("Your emoji mood: " + emoji)
```

```python
def emoji_ui(text):
    return gen_emojis(text)  # Make sure this function is defined above or imported

# 2️⃣ Create the Gradio interface
demo = gr.Interface(
    fn=emoji_ui,
    inputs=gr.Textbox(
        lines=2,
        placeholder="Type any sentence or phrase...",
        label="Enter your sentence"
    ),
    outputs=gr.Textbox(label="Suggested Emojis"),
    title="EmojiMood – Emoji Generator",
    description="Type an English or Hindi sentence and get matching emojis instantly! 😊"
)

# # 3️⃣ Launch the web app
# if __name__ == "__main__":
demo.launch()
```
